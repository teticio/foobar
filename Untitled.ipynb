````python
def solution(n):
    N = 20232
    sieve = [True] * N
    for i in range(3, int(N**0.5) + 1, 2):
        if sieve[i]:
            sieve[i * i::2 * i] = [False] * ((N - i * i - 1) // (2 * i) + 1)
    primes = '2' + ''.join(str(i) for i in range(3, N, 2) if sieve[i])
    return primes[n:n + 5]
````

````python
from random import random


def monte_carlo(matrix, sims):
    n = len(matrix)
    terminal_states = [s for s in range(n) if sum(matrix[s]) == 0]
    cum_probs = []
    for state in range(n):
        if state in terminal_states:
            cum_probs += [[]]
            continue
        denom = sum(matrix[state])
        cum_probs += [[prob / denom for prob in matrix[state]]]
        for s in range(1, n):
            cum_probs[state][s] += cum_probs[state][s - 1]
    state = start_state = 0
 #   print(f'{state}', end='')
    terminal_probs = [0] * len(terminal_states)
    terminations = 0
    for _ in range(sims):
        rnd = random()
        for next_state in range(n):
            if cum_probs[state][next_state] > rnd:
                break
        state = next_state
#        print(f'->{next_state}', end='')
        if state in terminal_states:
            terminal_probs[terminal_states.index(state)] += 1
            state = start_state
            terminations += 1
#            print()
#            print(f'{state}', end='')
    terminal_probs = [prob / terminations for prob in terminal_probs]
    return terminal_probs
````

````python
def transposeMatrix(m):
    return list(map(list, zip(*m)))

def getMatrixMinor(m,i,j):
    return [row[:j] + row[j+1:] for row in (m[:i]+m[i+1:])]

def getMatrixDeternminant(m):
    #base case for 2x2 matrix
    if len(m) == 2:
        return m[0][0]*m[1][1]-m[0][1]*m[1][0]

    determinant = 0
    for c in range(len(m)):
        determinant += ((-1)**c)*m[0][c]*getMatrixDeternminant(getMatrixMinor(m,0,c))
    return determinant

def getMatrixInverse(m):
    determinant = getMatrixDeternminant(m)
    #special case for 2x2 matrix:
    if len(m) == 2:
        return [[m[1][1]/determinant, -1*m[0][1]/determinant],
                [-1*m[1][0]/determinant, m[0][0]/determinant]]

    #find matrix of cofactors
    cofactors = []
    for r in range(len(m)):
        cofactorRow = []
        for c in range(len(m)):
            minor = getMatrixMinor(m,r,c)
            cofactorRow.append(((-1)**(r+c)) * getMatrixDeternminant(minor))
        cofactors.append(cofactorRow)
    n = len(cofactors)
    cofactors = transposeMatrix(cofactors)
    for r in range(len(cofactors)):
        for c in range(len(cofactors)):
            cofactors[r][c] = cofactors[r][c]/determinant
    return cofactors
````

````python
import numpy as np
from functools import reduce
from fractions import Fraction, gcd


def find_gcd(list):
    x = reduce(gcd, list)
    return x


def solution(matrix):
    n = len(matrix)
    terminal_states = [s for s in range(n) if sum(matrix[s]) == 0]
    new_matrix = []
    for i, row in enumerate(matrix):
        norm = sum(row)
        new_matrix += [[-float(element) / norm for element in row]] if norm else [[0] * len(row)]
        new_matrix[i][i] += 1
    inverse = np.linalg.inv(new_matrix)
    answer = [Fraction(inverse[0][s]).limit_denominator() for s in terminal_states]
    greatest_common_denominator = find_gcd(answer).denominator
    return [a.numerator * greatest_common_denominator // a.denominator for a in answer] + [greatest_common_denominator]
````

````python
import itertools


class Graph:
    def __init__(self, matrix):
        self.V = len(matrix)
        self.graph = []
        self.dist = [[]] * self.V

    def add_edge(self, s, d, w):
        self.graph.append([s, d, w])

    def bellmanFord(self, src):
        self.dist[src] = [float("Inf")] * self.V
        self.dist[src][src] = 0
        for temp in range(self.V - 1):
            for s, d, w in self.graph:
                if self.dist[src][s] != float(
                        "Inf") and self.dist[src][s] + w < self.dist[src][d]:
                    self.dist[src][d] = self.dist[src][s] + w
        for s, d, w in self.graph:
            if self.dist[src][s] != float(
                    "Inf") and self.dist[src][s] + w < self.dist[src][d]:
                return True  # graph contains negative cycle
        return False
    
    
def solution(matrix, time):
    n = len(matrix)
    g = Graph(matrix)
    for i in range(n):
        for j in range(n):
            if i == j:
                continue
            g.add_edge(i, j, matrix[i][j])
    negative_cycle = False
    for i in range(n):
        has_negative_cycle = g.bellmanFord(i)
        if has_negative_cycle:
            negative_cycle = True
    if negative_cycle:
        return list(range(0, len(matrix) - 2))
    
    best = 0
    bunnies = []
    for subset_size in range(1, n - 1): # not including bulkhead
        for subset in itertools.permutations(range(1, n - 1), subset_size): # not including bulkhead
            pos = 0
            cost = 0
            for next_pos in subset:
                cost += g.dist[pos][next_pos]
                pos = next_pos
            cost += g.dist[next_pos][n - 1]
            if cost <= time and len(subset) > best:
                best = len(subset)
                bunnies = [bunny - 1 for bunny in subset]
    bunnies.sort()
    return bunnies
````

````python
import math
import turtle

def bounce(pos, heading, width, height):
    radians = math.radians(heading)
    cos = math.cos(radians)
    sin = math.sin(radians)
    nearest = float("inf")
    if cos < 0: # left
        if (0 - pos[0]) / cos < nearest:
            nearest = (0 - pos[0]) / cos
            new_heading = 180 - heading
    elif cos > 0: # right
        if (width - pos[0]) / cos < nearest:
            nearest = (width - pos[0]) / cos
            new_heading = 180 - heading
    if sin < 0: # bottom
        if (0 - pos[1]) / sin < nearest:
            nearest = (0 - pos[1]) / sin
            new_heading = -heading
    elif sin > 0: # top
        if (height - pos[1]) / sin < nearest:
            nearest = (height - pos[1]) / sin
            new_heading = -heading
    new_pos = (pos[0] + nearest * cos, pos[1] + nearest * sin)
    return nearest, new_heading % 360, new_pos
````

````python
import math


def sign(x):
    return -1 if x < 0 else 1


def get_image(me, pos, dx, dy, width, height):
    if dx % 2 != 0:
        pos = (width - pos[0], pos[1])
    if dy % 2 != 0:
        pos = (pos[0], height - pos[1])
    pos = (pos[0] + dx * width, pos[1] + dy * height)
    dx = pos[0] - me[0]
    dy = pos[1] - me[1]
    angle = math.atan2(dy, dx)
    distance2 = dx * dx + dy * dy
    return angle, distance2


def solution(dims, me, target, distance):
    width = dims[0]
    height = dims[1]

    angles = {}
    for dx in range((me[0] - distance) // width - 1,
                    (me[0] + distance) // width + 1):
        for dy in range((me[1] - distance) // height - 1,
                        (me[1] + distance) // height + 1):
            if dx or dy:
                angle, distance2 = get_image(me, me, dx, dy, width, height)
                if angle not in angles:
                    angles[angle] = []
                angles[angle] += [(distance2, 'O')]
            angle, distance2 = get_image(me, target, dx, dy, width, height)
            if angle not in angles:
                angles[angle] = []
            angles[angle] += [(distance2, 'X')]

    hits = 0
    for angle in angles:
        angles[angle].sort()
        if angles[angle][0][0] <= distance * distance and angles[angle][0][
                1] == 'X':
            hits += 1
    return hits
````

The switching operation forms a group of all permutations. Any permutation of n elements leaves some elements fixed and permutes the others in a cycle of length equal to the number of non-fixed elements. Therefore, enumerating the cycle groups equates to enumerating the ways of summing integers to n.

This was too boring and required too much specialized knowledge of group theory. I based my solution on someone else's solution quite heavily.

````python
import math
import fractions


def ways_to_sum_to_n(n, N, floor=1):
    ways = [[int(_ == n - 1) for _ in range(N)]]
    for i in range(floor, n // 2 + 1):
        for j in ways_to_sum_to_n(n - i, N, i):
            ways += [[j[_] + int(_ == i - 1) for _ in range(N)]]
    return ways


def cycles(ways):
    n = math.factorial(len(ways))
    for i, j in enumerate(ways):
        n //= (i + 1)**j * math.factorial(j)
    return n


def solution(w, h, s):
    rows = ways_to_sum_to_n(h, h)
    columns = ways_to_sum_to_n(w, w)
    n = 0
    for row in rows:
        for column in columns:
            n += cycles(row) * cycles(column) * (s**sum([
                sum([
                    fractions.gcd(i + 1, j + 1) * f * g
                    for i, f in enumerate(row)
                ]) for j, g in enumerate(column)
            ]))
    return str(n // (math.factorial(w) * math.factorial(h)))
````
